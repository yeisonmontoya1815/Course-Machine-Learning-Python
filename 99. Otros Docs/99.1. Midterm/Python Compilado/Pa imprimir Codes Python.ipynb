# 2.1. for looping common operations

```python
# Loop through a list of numbers
numbers = [1, 2, 3, 4, 5]
print("Loop through a list of numbers:")
for number in numbers:
    print(number)

# Loop through a range of numbers
print("\nLoop through a range of numbers:")
for i in range(5):
    print(i)

# Loop through characters in a string
print("\nLoop through characters in a string:")
for char in "Python":
    print(char)

# Loop through key-value pairs in a dictionary
print("\nLoop through key-value pairs in a dictionary:")
person = {'name': 'John', 'age': 30, 'city': 'New York'}
for key, value in person.items():
    print(key, value)

# Loop through elements with index using enumerate
print("\nLoop through elements with index using enumerate:")
fruits = ['apple', 'banana', 'cherry']
for index, fruit in enumerate(fruits):
    print(index, fruit)

# Nested loops
print("\nNested loops:")
for i in range(3):
    for j in range(3):
        print(i, j)

# Continue statement to skip iteration
print("\nContinue statement to skip iteration:")
numbers = [1, 2, 3, 4, 5]
for number in numbers:
    if number == 3:
        continue  # Skip the rest of the loop for this iteration
    print(number)

# Loop with else statement
print("\nLoop with else statement:")
for i in range(5):
    print(i)
else:
    print("Loop completed without a break.")
```

# 2.2. if - else statements

```python
# Example: Check conditions and perform actions accordingly

# Scenario 1: Numeric Comparison
x = 10
if x > 5:
    print("x is greater than 5")
elif x == 5:
    print("x is equal to 5")
else:
    print("x is less than 5")

# Scenario 2: String Comparison
country = "USA"
if country == "USA":
    print("Country is United States")
elif country == "Canada":
    print("Country is Canada")
else:
    print("Country is neither USA nor Canada")

# Scenario 3: Checking for Membership
fruits = ['apple', 'banana', 'orange']
fruit_to_check = 'banana'
if fruit_to_check in fruits:
    print(f"{fruit_to_check} is in the list")
else:
    print(f"{fruit_to_check} is not in the list")

# Scenario 4: Logical Conditions
age = 25
income = 50000
if age > 18 and income > 30000:
    print("Eligible for loan")
else:
    print("Not eligible for loan")

# Scenario 5: Nested if-else
temperature = 28
if temperature > 30:
    print("High temperature detected")
else:
    print("Normal or low temperature detected")
    if temperature < 0:
        print("Extreme cold temperature detected")

# Scenario 6: Ternary Operator
x = 10
message = "Even" if x % 2 == 0 else "Odd"
print(message)

# Scenario 7: Checking for None
value = None
if value is None:
    print("Value is None")
else:
    print("Value is not None")

# Scenario 8: Using not in conditions
is_raining = False
if not is_raining:
    print("No need for an umbrella")
else:
    print("Bring an umbrella")

# Scenario 9: Multiple conditions using or
day = "Saturday"
if day == "Saturday" or day == "Sunday":
    print("It's the weekend!")
else:
    print("It's a weekday")

# Scenario 10: Multiple conditions using and
temperature = 25
if temperature > 20 and temperature < 30:
    print("Temperature is comfortable")
else:
    print("Temperature is not in the comfort range")
```

# 2.3. While loop over conditions

```python
# Simple While Loop:
print("Simple While Loop:")
count = 0
while count < 5:
    print(count)
    count += 1

# Using else with While Loop:
print("\nUsing else with While Loop:")
count = 0
while count < 5:
    print(count)
    count += 1
else:
    print("Loop completed without a break.")

# While Loop with continue:
print("\nWhile Loop with continue:")
count = 0
while count < 5:
    count += 1
    if count == 3:
        continue  # Skip the rest of the loop for this iteration
    print(count)

# Using break to Exit Loop Prematurely:
print("\nUsing break to Exit Loop Prematurely:")
count = 0
while count < 5:
    if count == 3:
        break  # Exit the loop when count is 3
    print(count)
    count += 1

# While Loop with Condition:
print("\nWhile Loop with Condition:")
temperature = 3
while temperature < 10:
    print(f"Temperature is {temperature}°C. Heating...")
    temperature += 1

# Using else with a Condition:
print("\nUsing else with a Condition:")
temperature = 0
while temperature < 100:
    print(f"Temperature is {temperature}°C. Heating...")
    temperature += 10
else:
    print("Reached maximum temperature.")
```

## **working from array dataframe**

```python
# Import the necessary libraries
import pandas as pd

# creting the Array DataFrame
data = {
    'cars_per_cap': [809, 731, 588, 18, 200, 70, 45],
    'country': ['United States', 'Australia', 'Japan', 'India', 'Russia', 'Morocco', 'Egypt'],
    'drives_right': [True, False, False, False, True, True, True]
}

cars = pd.DataFrame(data)
# Display the original DataFrame
print("Original DataFrame:")
print(cars)

# Drop rows with NaN values, if any
cars_cleaned = cars.dropna()
# Display the DataFrame after dropping rows with NaN (in this case, there shouldn't be any)
print("\nDataFrame after dropping rows with NaN:")
print(cars_cleaned)

# Drop duplicate rows, if any
cars_cleaned = cars_cleaned.drop_duplicates()
# Display the DataFrame after dropping duplicate rows (in this case, there shouldn't be any)
print("\nDataFrame after dropping duplicate rows:")
print(cars_cleaned)

# Drop rows with duplicate values in the 'country' column
cars_cleaned = cars_cleaned.drop_duplicates(subset='country')
# Display the DataFrame after dropping rows with duplicate values based on 'country'
print("\nDataFrame after dropping rows with duplicate values in 'country':")
print(cars_cleaned)

# Reset the index of the DataFrame after cleaning operations
cars_cleaned.reset_index(drop=True, inplace=True)
# Display the DataFrame with the reset index
print("\nDataFrame with the reset index:")
print(cars_cleaned)

# Print the first 3 observations in the DataFrame
print("\nFirst 3 observations in the DataFrame:")
print(cars_cleaned.head(3))
# Print observations 4, 5, and 6 in the DataFrame
print("\nObservations 4, 5, and 6 in the DataFrame:")
print(cars_cleaned[3:6])

# Print the observation for Japan using label-based indexing (loc)
print("\nObservation for Japan using loc:")
print(cars_cleaned.loc[cars_cleaned['country'] == 'Japan'])

# Print observations for Australia and Egypt using label-based indexing
print("\nObservations for Australia and Egypt using loc:")
print(cars_cleaned.loc[cars_cleaned['country'].isin(['Australia', 'Egypt'])])

# Print the 'drives_right' value for Morocco using label-based indexing
print("\n'Drives_right' value for Morocco using loc:")
print(cars_cleaned.loc[cars_cleaned['country'] == 'Morocco', 'drives_right'])

# Print a sub-DataFrame for Russia and Morocco with selected columns
print("\nSub-DataFrame for Russia and Morocco with selected columns:")
print(cars_cleaned.loc[cars_cleaned['country'].isin(['Russia', 'Morocco']), ['country', 'drives_right']])

# Print the 'drives_right' column as a Pandas Series
print("\n'Drives_right' column as a Pandas Series:")
print(cars_cleaned['drives_right'])

# Print the 'drives_right' column as a Pandas DataFrame
print("\n'Drives_right' column as a Pandas DataFrame:")
print(cars_cleaned[['drives_right']])

# Print 'cars_per_cap' and 'drives_right' columns as a DataFrame
print("\n'cars_per_cap' and 'drives_right' columns as a DataFrame:")
print(cars_cleaned[['cars_per_cap', 'drives_right']])
```

# working with functions

```python

import pandas as pd

# Create a DataFrame
df = pd.DataFrame({'lang': ['Python', 'Java', 'Python', 'C++', 'Python']})


# Define count_entries()
def count_entries(df, col_name='lang'):
    """Return a dictionary with counts of
    occurrences as value for each key."""
    
    # Raise a ValueError if col_name is NOT in DataFrame
    if col_name not in df.columns:
        raise ValueError('The DataFrame does not have a '+col_name+' column.')

    # Initialize an empty dictionary: cols_count
    cols_count = {}
    
    # Extract column from DataFrame: col
    col = df[col_name]
    
    # Iterate over the column in DataFrame
    for entry in col:

        # If entry is in cols_count, add 1
        if entry in cols_count.keys():
            cols_count[entry] += 1
            # Else add the entry to cols_count, set the value to 1
        else:
            cols_count[entry] = 1
        
        # Return the cols_count dictionary
    return cols_count

# Call count_entries(): result1
result1 = count_entries(df)

# Print result1
print(result1)
```
